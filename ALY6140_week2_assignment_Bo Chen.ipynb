```python
#Create a function to read the txt file
def import_text(file_path):
    """
    This function reads a text file and returns its contents as a list of lines.

    :param file_path: str, the path to the text file
    :return: list, the contents of the text file
    """
    with open(file_path, 'r') as file:
        lines = file.readlines()
    return lines
if __name__ == '__main__':
    text_path = '/content/drive/MyDrive/Colab Notebooks/ALY6140 assignment folder/Week2/network_data.txt'

    #Call the functions and store the results
    text_data = import_text(text_path)
```

```python
#Create a function to read the excel file,exlude the blank rows and coloumns
def import_excel(file_path, sheet_name):
    """
    This function reads an Excel file and returns a specified sheet's contents as a DataFrame.

    :param file_path: str, the path to the excel file
    :param sheet_name: str, the name of the sheet to be imported
    :return: DataFrame, the contents of the specified sheet
    """
    df = pd.read_excel(file_path, sheet_name=sheet_name,header=24,usecols=lambda x: 'Unnamed' not in x,)
    return df
if __name__ == '__main__':
    excel_path = '/content/drive/MyDrive/Colab Notebooks/ALY6140 assignment folder/Week2/Excel_report.xlsx'

    #Call the functions and store the results
    excel_data = import_excel(excel_path,'financials')
```
